```python
# CADL4: Topic Modeling with LDA
import gensim
from gensim import corpora
from nltk.corpus import stopwords
import nltk
import pyLDAvis
import pyLDAvis.gensim_models as gensimvis

# Download NLTK stopwords
nltk.download('stopwords')
```

```python
# Sample text corpus
corpus = [
    "AI and machine learning are revolutionizing healthcare.",
    "Renewable energy sources are key to fighting climate change.",
    "The stock market fluctuates based on economic and political news.",
    "Yoga and meditation are becoming popular for mental health.",
    "Mars missions are advancing space exploration and technology."
]
```

```python
# Stopwords
stop_words = set(stopwords.words('english'))

# Clean and tokenize
texts = [
    [word.lower() for word in doc.split() if word.lower() not in stop_words]
    for doc in corpus
]

print("Preprocessed Texts:")
for i, text in enumerate(texts):
    print(f"Doc {i+1}: {text}")
```

```python
# Create dictionary
dictionary = corpora.Dictionary(texts)

# Create Bag of Words (BOW) corpus
doc_term_matrix = [dictionary.doc2bow(text) for text in texts]

# Print dictionary and BOW
print("\nDictionary:")
print(dictionary.token2id)

print("\nDocument-Term Matrix (first document):")
print(doc_term_matrix[0])
```

```python
# LDA Model
lda_model = gensim.models.LdaModel(
    corpus=doc_term_matrix,
    num_topics=3,        # Number of topics
    id2word=dictionary,
    passes=15,           # Training passes
    random_state=42
)

# Print discovered topics
print("\n🔹 Topics Identified:")
for idx, topic in lda_model.print_topics(num_words=5):
    print(f"Topic {idx}: {topic}")
```

```python
# Prepare pyLDAvis visualization
pyLDAvis.enable_notebook()
vis = gensimvis.prepare(lda_model, doc_term_matrix, dictionary)
vis
```
